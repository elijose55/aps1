```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint

t = np.arange(0,5,0.01)

r=2.05   #comprimento da corda
m=0.055  #massa da bola
g=9.8
ap = math.radians(30)  #angulo inicial pequeno
am = math.radians(60)  #angulo  inicial grande


def EqDif(P,t): #equacao diferencial do angulo e velocidade angular
    ùúÉ=P[0]
    ùúî=P[1]
    dùúÉdt = ùúî
    dùúîdt = (-g*(math.sin(ùúÉ))/r)
    return [dùúÉdt, dùúîdt]

# modelagem do angulo menor        

w0=[am, 0]
             
res = odeint(EqDif, w0, t)

x = []
y = []

for ùúÉ in res[:,0]:
    x.append(r*-math.cos(ùúÉ))
    
for ùúÉ in res[:,0]:
    y.append(r*math.sin(ùúÉ))
    
plt.title('Trajet√≥ria do p√™ndulo - 30¬∞')
plt.axis('equal')
plt.plot(y,x)
plt.ylabel('metros')
plt.xlabel('metros')
plt.show()

#modelagem do angulo maior

w1=[ap, 0]
res1 = odeint(EqDif, w1, t)


x = []
y = []

for ùúÉ in res1[:,0]:
    x.append(r*-math.cos(ùúÉ))
    
for ùúÉ in res1[:,0]:
    y.append(r*math.sin(ùúÉ))
    
plt.title('Trajet√≥ria do p√™ndulo - 60¬∞')
plt.axis('equal')
plt.ylabel('metros')
plt.xlabel('metros')
plt.plot(y,x)
plt.show()
```
