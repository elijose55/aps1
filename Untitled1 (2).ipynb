```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint

t = np.arange(0,5,0.01)

r=2.05   #comprimento da corda
m=0.055  #massa da bola
g=9.8
ap = math.radians(30)  #angulo inicial pequeno
am = math.radians(60)  #angulo  inicial grande


def EqDif(P,t): #equacao diferencial do angulo e velocidade angular
    𝜃=P[0]
    𝜔=P[1]
    d𝜃dt = 𝜔
    d𝜔dt = (-g*(math.sin(𝜃))/r)
    return [d𝜃dt, d𝜔dt]

# modelagem do angulo menor        

w0=[am, 0]
             
res = odeint(EqDif, w0, t)

x = []
y = []

for 𝜃 in res[:,0]:
    x.append(r*-math.cos(𝜃))
    
for 𝜃 in res[:,0]:
    y.append(r*math.sin(𝜃))
    
plt.title('Trajetória do pêndulo - 30°')
plt.axis('equal')
plt.plot(y,x)
plt.ylabel('metros')
plt.xlabel('metros')
plt.show()

#modelagem do angulo maior

w1=[ap, 0]
res1 = odeint(EqDif, w1, t)


x = []
y = []

for 𝜃 in res1[:,0]:
    x.append(r*-math.cos(𝜃))
    
for 𝜃 in res1[:,0]:
    y.append(r*math.sin(𝜃))
    
plt.title('Trajetória do pêndulo - 60°')
plt.axis('equal')
plt.ylabel('metros')
plt.xlabel('metros')
plt.plot(y,x)
plt.show()
```

```python
#grafico do angulo do pendulo

plt.title('Ângulo do pêndulo por tempo - 30°')
plt.axis('equal')
plt.grid(True)
plt.ylabel('ângulo (rad)')
plt.xlabel('tempo (seg)')
plt.plot(t, res[:,0])
plt.show()

plt.title('Ângulo do pêndulo por tempo - 60°')
plt.axis('equal')
plt.grid(True)
plt.plot(t, res1[:,0])
plt.ylabel('ângulo (rad)')
plt.xlabel('tempo (seg)')
plt.show()

periodo=2*math.pi*(r/g)**(1/2)
print('P =', periodo)
amplitude = max(res[:,0])
print('A =', amplitude)
amplitude1 = max(res1[:,0])
print('A =', amplitude1)
```
